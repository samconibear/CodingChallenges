```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

stock = 'TSLA'
def current(stock):
    # Returns the current share price (live up to ~20 seconds) and a timestamp.
    # in the form (878.5700073242188, Timestamp('2021-01-27 14:27:18-0500', tz='America/New_York'))
    stock = yf.Ticker(stock)
    c = stock.history(period='1d',interval='15m')
    currentPrice = c['Close'].iloc[-1]
    currentTime = c.index[-1]
    return currentPrice, currentTime


def stockdata(stock,Period='3mo',Interval='1h'):
    stock = yf.Ticker(stock)
    history = stock.history(period=Period,interval=Interval)
    return history['Close']

fig, ax = plt.subplots()  
ax.plot(stockdata(stock))
plt.show()
```
